```python
# Pick the data for male OR female and fit the income vs. education using a linear equation.

# We will go with males. First, I need need to import the correct libraries.
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Then, I'm going to input the data into an array.

Income_arr = np.array([36700, 46400, 50200, 54200, 75100, 89200]) # these are the levels of income given by the website organized in an array.
Education_arr = np.arange(len(Income_arr)) # Counts how many values are in the Income_arr (5) and organizes it in an array.
plt.plot(Education_arr, Income_arr, "o-")

# Now we plot, labeling the axis' and showing it.
plt.xlabel("Education")
plt.ylabel("Income[$]")
plt.title("Income vs. Education")
plt.show()
```

```python
# Now, I am going to label each number on the x axis according to their education level. Instead of 1-5, it now puts these phrases
# in their place.

Education_levels = np.array(["Less than highschool completion", "Highschool completion", "some college but no degree",
                             "associate's degree", "bachelor's degree", "master's degree or higher"])

# Now lets replot given our new labels.

plt.plot(Education_arr, Income_arr, "o-")
plt.xticks(Education_arr, Education_levels)
plt.xlabel("Education")
plt.ylabel("Income [$]")
plt.xticks(Education_arr, Education_levels, rotation=45, ha="right") # Tilts the labels 45 degrees.
plt.show()
```

```python
# I want to show the linear relationship between the points with aline best fit. To do this
# I'm going to use polyfit, considering the education and income arrays mentioned previously.

z_linear = np.polyfit(Education_arr,Income_arr,1)
C_linear = z_linear[0] * Education_arr + z_linear[1] # Performs a linear combination of the Education_arr array.

# Reploting:

Income_arr = np.array([35500.,41800.,45200.,49500.,66600.,80200.])
Education_arr = np.arange(len(Income_arr))
plt.plot(Education_arr, Income_arr, "o-")
plt.plot(Education_arr,C_linear, color = "red", label = "Linear fit")
plt.xticks(Education_arr, Education_levels, rotation=45, ha="right")
plt.legend() # Adds a legend to tell the reader what the linear fit of the graph is.
plt.xlabel("Education Level")
plt.ylabel("Income [$]")
plt.show()
```
